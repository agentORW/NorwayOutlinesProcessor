# Step 2
Now we are just going to crop the GeoJson data to get a spesific area.

## Imports

```python
import geopandas
```

## Define bounding box

```python
min_coordinates = list(map(float, "60.37656945862301, 5.298660940936371".split(", "))) # Lower left
max_coordinates = list(map(float, "60.40573082269235, 5.349207903976764".split(", "))) # Upper right

selected_bbox = (min_coordinates[1], min_coordinates[0], max_coordinates[1], max_coordinates[0])
```

## Crop building data

```python
layer = "fkb_bygning_grense"
input_file_path = f"geojson/{layer}.geojson"
output_file_path = f"cropped/{layer}-CROPPED.geojson"

data = geopandas.read_file(input_path, bbox=selected_bbox)
data.to_file(output_file_path)
```

## Crop road data

```python
layers = ["fkb_veg_senterlinje", "fkb_veg_grense", "fkb_vek_omr√•de"]
input_file_paths = [f"geojson/{layer}.geojson" for layer in layers]
output_file_paths = [f"cropped/{layer}-CROPPED.geojson" for layer in layers]

for index, input_path in enumerate(input_file_paths):
    data = geopandas.read_file(input_path, bbox=selected_bbox)
    data.to_file(output_file_paths[index])
```
