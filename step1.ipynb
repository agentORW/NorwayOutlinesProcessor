# Step 1
We need to first convert the GML to GeoJson. Were going to use ogr2ogr for this

```python
import subprocess
```

## Building data
We only care about one layer here

```python
input_file = "gml\\Basisdata_0000_Norge_5973_FKB-Bygning_FGDB\\Basisdata_0000_Norge_5973_FKB-Bygning_FGDB.gdb"
layer = "fkb_bygning_grense"
output_file = f"geojson\\{layer}.geojson"
command = f'ogr2ogr -t_srs EPSG:4326 "GeoJson" {output_file} {input_file} {layer} -nlt multipolygonz'
subprocess.run(command, shell=True, check=True)
```

## Road data
Here we have multiple layers

```python
input_file = "gml\\Basisdata_0000_Norge_5973_FKB-Veg_FGDB\\Basisdata_0000_Norge_5973_FKB-Veg_FGDB.gdb"
layers = ["fkb_veg_senterlinje", "fkb_veg_grense", "fkb_vek_omr√•de"]
output_file_paths = ["geojson\\"+layer+".geojson" for layer in layers]

for index, layer in enumerate(layers):
    command = f'ogr2ogr -t_srs EPSG:4326 "GeoJson" {output_file_paths[index]} {input_file} {layer} -nlt multipolygonz'
    subprocess.run(command, shell=True, check=True)
```

These should all now be saved in a `geojson` folder
